Importando os arquivos de dados de treino e teste

```python
import pandas as pd

url_treino = 'https://raw.githubusercontent.com/safira1344/classificador-neural-diabetes/refs/heads/main/dados_treino.csv'
dados_treino = pd.read_csv(url_treino,sep=',', encoding = 'latin1')
url_teste = 'https://raw.githubusercontent.com/safira1344/classificador-neural-diabetes/refs/heads/main/dados_teste.csv'
dados_teste = pd.read_csv(url_teste,sep=',', encoding = 'latin1')
```

Separando os dados de input de entrada dos dados de resultados, tanto para base de treino quanto para a base de teste


```python
x_treino = dados_treino.drop('Outcome', axis=1)
y_treino = dados_treino['Outcome']

x_teste = dados_teste.drop('Outcome', axis=1)
y_teste = dados_teste['Outcome']
```

Normalizando os dados de input

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_treino_normalizado = scaler.fit_transform(x_treino)
x_teste_normalizado = scaler.transform(x_teste)
```

Treinando o modelo usando o algoritmo de Perceptron

```python
from sklearn.linear_model import Perceptron

modelo = Perceptron()
modelo.fit(x_treino_normalizado, y_treino)
```

Testando acurácia

```python
print('Acurácia: %.3f' % modelo.score(x_teste_normalizado, y_teste))
```

Realizando a predição com os dados de teste

```python
base_Predicao = modelo.predict((x_teste_normalizado))

comparacao = pd.DataFrame({
    'Real': y_teste.values,
    'Predito': base_Predicao
})

display(comparacao)
```
